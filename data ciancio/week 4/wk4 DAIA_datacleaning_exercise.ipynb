> alles hierboven is van het original document <p>hieronder verander ik de code zodat het in een copie van de original data frame.

```python
df_repaired = df.copy()
df_repaired['Sector'] = df_repaired['Sector'].str.strip()
# we can also include the newly split columns in the dataframe itself
df_repaired[['sector1', 'sector2', 'sector3']] = df_repaired['Sector'].str.split(',', expand = True)
# and delete the original column to clean the set (and prevent duplicate values)
df_repaired.drop('Sector', axis='columns', inplace=True)
df_repaired.head(10)
```

```python
# By checking the unique values we actually find duplicated values in 'Status of deal'
df['Status of deal'].unique()
```

```python
# among the string functions is strip() to remove all white spaces (and other 'invisible characters' )
df_repaired['Status of deal'] = df_repaired['Status of deal'].str.strip()
df_repaired['Status of deal'].unique()
```

```python
# check for duplicate values
df.duplicated().value_counts()
```

> hieronder zorg ik ervoor dat ik de text aanpas zodat het meeste van de informatie in getallen word neergezet. <p>het is me niet voledig gelukt om alles in getallen te krijgen. <p>dit houd ook in dat ik kijk naar een minimaal investering en een maximaale investiring.

```python
# try to 'fix' US$77 million in the Projected investment column
# to be continued...
million_mask = df_repaired['Projected investment'].str.contains('million', na=False)

df_repaired['Projected investment'] = df_repaired['Projected investment'].replace({'US\$': '', ' million' : ''}, regex=True)

range_mask = df_repaired['Projected investment'].str.contains('-', na=False)

temp_test = df_repaired['Projected investment'].str.split('-', expand=True)

temp_test.columns = ['Min_Investment', 'Max_Investment']

df_repaired.loc[range_mask, ['Min_Investment', 'Max_Investment']] = temp_test

single_value_mask = ~range_mask

# For rows with a single value, use the same value for both minimum and maximum investment
df_repaired.loc[single_value_mask, 'Min_Investment'] = df_repaired.loc[single_value_mask, 'Projected investment']
df_repaired.loc[single_value_mask, 'Max_Investment'] = df_repaired.loc[single_value_mask, 'Projected investment']

display(df_repaired.isna().sum())
display(df_repaired.sort_values(by='Min_Investment').head(10))
```
